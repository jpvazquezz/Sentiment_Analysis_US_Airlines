### Data Cleaning

As this is an supervised machine learning task, it is required to explictly designed the labels to classify the input data that our model would use. In our case, the sentiment analysis involves the labels 'positive', 'neutral' and 'negative' that are assigned to the variable `labels` below. the label set consists of the sentiments of the tweets we have to predict. Our features are the tweets themselves, and are assigned to the variables `features`.

```python
features = airlines['text'].values
labels = airlines['airline_sentiment'].values
```

Below, a function to process the data is created and aimes to clean and remove all the unnecessary elements in our text features. For this task, we use regular expressions to nicely manage to alter the tweets.

```python
def preprocess(features):
    processed_features = []
    for sentence in range(0,len(features)):
        processed_feature = re.sub(r'\W',' ',str(features[sentence]))    # Remove all the special characters
        processed_feature = re.sub(r'\s+[a-zA-z]\s+',' ',processed_feature)    # remove all single characters
        processed_feature = re.sub(r'\^[a-zA-z]\s+',' ', processed_feature)    # Remove single characters from the start
        processed_feature = re.sub(r'\s+', ' ',processed_feature,flags=re.I)    # Substituting multiple spaces with single space
        processed_feature = re.sub(r'^b\s+','',processed_feature)     # Removing prefixed 'b'
        processed_feature = processed_feature.lower()    # Convert to lowercase
        processed_features.append(processed_feature)
    return processed_features
```

```python
processed_features = preprocess(features)
processed_features
```

### Modelling

To make machine learning algorithms work with text, we first have to convert text to numbers. To do so, we'll use the TD-IDF method (Term frequency - Inverse document frequency). The TF-IDF measures how relevant a word is in a document in a collection of documents, which in our case are the tweets. This statistical measure is a result of the multiplication of the term frequency (the number of times a word appears in a document - tweet) and the inverse document frequency (a rarity measure - the higher the less frequent a term appears in all documents).

For this purpose, we use Python's Scikit-Learn library contains the `TfidfVectorizer` class that can be used to convert text features into TF-IDF feature vectors. We specify that the `max_features` should be 2500, meaning that algorithm only uses the 2500 most frequently occurring words to create the feature vector. `max_df` defines that only the words that occur in a maximum of 80% of the documents should be used as `min-df` is set to 7 which only include words that occur in at least 7 documents.

```python
vec = TfidfVectorizer(max_features=2500, min_df=7,max_df = 0.8,stop_words=stopwords.words('english'))

processed_features = vec.fit_transform(processed_features).toarray()
```

Now, we need to divide the dataset into training and testing sets. The training set will be used to train the algorithm while the test set will be used to evaluate the performance of the machine learning model. For this, we use the the `train_test_split` class from the sklearn.model_selection module.

The method requires the features set as the first parameter, the labels set as the second paramenter as a value for the `test_size`, which means that the division between training and testing sets is defined by a percentage, in our case, 20% for testing and 80% for training.

```python
X_train, X_test,y_train, y_test = train_test_split(processed_features, labels, test_size=0.2, random_state=0)
```

We'll use two different machine learning algorithms: Random Forest Classifier and Support Vector Machine (SVM).

### Random Forest Classifier

Random forests are an ensemble learning, which is a type of learning where you join different types of algorithms or same algorithm multiple times to form a more powerful prediction model. It operates by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes, in the classification case.

```python
classifier = RandomForestClassifier(n_estimators=200, random_state=0)
classifier.fit(X_train,y_train)
```

```python
predict = classifier.predict(X_test)
```

### Support Vector Machine (SVM)

SVM is a type of supervised machine learning classification algorithm. SVM selects the decision boundary that maximizes the distance from the nearest data points of all the classes, finding the most optimal decision boundary.

```python
svcclassifier = SVC()
svcclassifier.fit(X_train,y_train)
```

```python
prediction = svcclassifier.predict(X_test)
```

### Evalution

For the evalution of our models' perfomance we use the F1-score and, mainly, the accuracy score.
F1-score is a measure of a test's accuracy calculated from the precision and recall of the test, where the precision is the number of correctly identified positive results divided by the number of all positive results, including those not identified correctly, and the recall is the number of correctly identified positive results divided by the number of all samples that should have been identified as positive.

The accuracy score computes the subset accuracy in multilabel classification.

```python
# Random Forest F1-score
print(classification_report(y_test,predict))
```

```python
# SVM F1-score
print(classification_report(y_test,prediction))
```

```python
# Random Forest Accuracy score
print(accuracy_score(y_test, predict))
```

```python
# SVM Accuracy score
print(accuracy_score(y_test, prediction))
```

From these results, we can conclude that the Support Vector Machine had a sliglty better perfomance and, therefore, is more recommended for the role of implementing a sentiment analysis regarding the investigation of the tweets mentions of US airlines.
